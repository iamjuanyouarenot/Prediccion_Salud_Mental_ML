```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import joblib
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# --- 1. Carga del Dataset ---
print("--- 1. Carga del Dataset ---")

from google.colab import files
import io

print("Por favor, sube tu archivo 'survey.csv'.")
uploaded = files.upload()

try:
    filename = list(uploaded.keys())[0]
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
    print(f"Archivo '{filename}' cargado exitosamente.")
except IndexError:
    print("Error: No se ha subido ningún archivo. Generando un dataset ficticio para la demostración.")
    # --- Generación de Dataset Ficticio para Demostración (si no se sube un archivo) ---
    data_size = 5000
    np.random.seed(42)

    data = {
        'Age': np.random.randint(18, 65, data_size),
        'Gender': np.random.choice(['Male', 'Female', 'Other'], data_size),
        'Country': np.random.choice(['USA', 'Canada', 'UK', 'Australia'], data_size),
        'state': np.random.choice(['IL', 'CA', 'TX', np.nan], data_size), # Simular nulos
        'self_employed': np.random.choice(['Yes', 'No'], data_size),
        'family_history': np.random.choice(['Yes', 'No'], data_size), # Proxy feature
        'treatment': np.random.choice(['Yes', 'No'], data_size), # Proxy feature
        'work_interfere': np.random.choice(['Never', 'Rarely', 'Sometimes', 'Often', np.nan], data_size), # Proxy feature, simulate NA
        'no_employees': np.random.choice(['1-5', '6-25', '26-100', '101-500', '501-1000', 'More than 1000'], data_size),
        'remote_work': np.random.choice(['Yes', 'No'], data_size),
        'tech_company': np.random.choice(['Yes', 'No'], data_size),
        'benefits': np.random.choice(['Yes', 'No', "Don't know"], data_size),
        'care_options': np.random.choice(['Yes', 'No', 'Not sure'], data_size),
        'wellness_program': np.random.choice(['Yes', 'No', "Don't know"], data_size),
        'seek_help': np.random.choice(['Yes', 'No', "Don't know"], data_size), # Proxy feature
        'anonymity': np.random.choice(['Yes', 'No', "Don't know"], data_size),
        'leave': np.random.choice(['Somewhat easy', 'Very easy', 'Somewhat difficult', 'Very difficult', "Don't know"], data_size),
        'mental_health_consequence': np.random.choice(['Yes', 'No', 'Maybe'], data_size), # Proxy feature
        'phys_health_consequence': np.random.choice(['Yes', 'No', 'Maybe'], data_size),
        'coworkers': np.random.choice(['Yes', 'No', 'Some of them'], data_size),
        'supervisor': np.random.choice(['Yes', 'No', 'Some of them'], data_size),
        'mental_health_interview': np.random.choice(['Yes', 'No', 'Maybe'], data_size),
        'phys_health_interview': np.random.choice(['Yes', 'No', 'Maybe'], data_size),
        'mental_vs_physical': np.random.choice(['Yes', 'No', "Don't know"], data_size),
        'obs_consequence': np.random.choice(['Yes', 'No'], data_size),
        'comments': [f'comment_{i}' if i % 100 == 0 else np.nan for i in range(data_size)], # Simular nulos
        'Timestamp': pd.to_datetime(pd.Series(pd.date_range('2023-01-01', periods=data_size))),
    }
    df = pd.DataFrame(data)
    # --- Fin de la Generación de Dataset Ficticio ---

print(f"\nDimensiones iniciales del dataset: {df.shape}")
print("Primeras 5 filas del dataset:")
print(df.head())
print("\nInformación del dataset (primeras filas para tipos de datos):")
df.info()

# --- 2. Preprocesamiento y Derivación de Variable Objetivo ---
print("\n--- 2. Preprocesamiento y Derivación de Variable Objetivo ---")

# 2.1 Eliminación de columnas innecesarias
columns_to_drop = ['Timestamp', 'state', 'comments']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')
print(f"Columnas eliminadas (si existen): {', '.join([col for col in columns_to_drop if col not in df.columns])}")

# 2.2 Derivación de la Variable Objetivo: 'Mental_Health_Concern' (binaria)
# Basada en las columnas proxy disponibles en tu dataset
def derivar_preocupacion_salud_mental(row):
    # Condición 1: Búsqueda de tratamiento
    if row['treatment'] == 'Yes':
        return 1
    # Condición 2: Interferencia del trabajo debido a salud mental (no "Never" ni NaN)
    if pd.notna(row['work_interfere']) and row['work_interfere'] not in ['Never', 'NA']:
        return 1
    # Condición 3: Consecuencias de salud mental
    if row['mental_health_consequence'] == 'Yes':
        return 1
    # Condición 4: Búsqueda de ayuda
    if row['seek_help'] == 'Yes':
        return 1
    # Condición 5: Historial familiar de enfermedad mental
    if row['family_history'] == 'Yes':
        return 1
    return 0 # Si ninguna de las condiciones anteriores se cumple

# Convertir columnas a tipo object para asegurar el mapeo de strings si son de tipo 'object'
# Algunas columnas pueden venir con tipo 'object' pero tener valores como 'Yes', 'No', 'NA' (string)
# Asegurar que los valores de las columnas proxy sean tratados como strings antes de aplicar la función.
proxy_cols = ['treatment', 'work_interfere', 'mental_health_consequence', 'seek_help', 'family_history']
for col in proxy_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).replace('nan', np.nan) # Convertir 'nan' string a NaN real si es necesario

df['Mental_Health_Concern'] = df.apply(derivar_preocupacion_salud_mental, axis=1)
print("Columna 'Mental_Health_Concern' creada (0: No preocupación, 1: Preocupación).")

# 2.3 Manejo de datos faltantes (general, después de derivación)
# Después de todas las transformaciones, se eliminan filas con cualquier NaN residual
print(f"\nFilas antes de eliminar nulos: {df.shape[0]}")
df.dropna(inplace=True)
print(f"Filas después de eliminar nulos: {df.shape[0]}")

# 2.4 Codificación de variables categóricas (incluyendo las proxy si no son binarias 0/1)
print("\nCodificando variables categóricas...")
# Excluir la variable objetivo 'Mental_Health_Concern'
cols_to_exclude = ['Mental_Health_Concern']
categorical_cols = df.drop(columns=cols_to_exclude, errors='ignore').select_dtypes(include=['object']).columns

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    print(f"Columna '{col}' codificada a valores numéricos.")

print("\nDataset después del preprocesamiento completo:")
print(df.head())
print(f"\nConteo de la variable objetivo 'Mental_Health_Concern':\n{df['Mental_Health_Concern'].value_counts()}")

# Visualización de la Distribución de la Variable Objetivo
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Mental_Health_Concern", palette='viridis')
plt.title("Distribución de Preocupación por Salud Mental")
plt.xlabel("Nivel (0: No Preocupación, 1: Preocupación)")
plt.ylabel("Cantidad")
plt.xticks([0, 1], ['No Preocupación', 'Preocupación'])
plt.tight_layout()
plt.show()

# --- 3. Definición de Características (X) y Variable Objetivo (y) ---
print("\n--- 3. Definición de Características (X) y Variable Objetivo (y) ---")

# X incluirá todas las columnas excepto la variable objetivo
X = df.drop(columns=['Mental_Health_Concern'], errors='ignore')
y = df["Mental_Health_Concern"]

# La variable objetivo ya es numérica (0 o 1), no necesita LabelEncoder adicional para y_encoded
y_encoded = y.values
target_names = ['No Preocupación', 'Preocupación']
print(f"Niveles de Preocupación por Salud Mental: {target_names}")

# División del conjunto de datos en entrenamiento y prueba (80%-20%)
# stratify=y_encoded para asegurar que la proporción de clases sea similar en ambos conjuntos
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print(f"\nDimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
print(f"Dimensiones de y_test: {y_test.shape}")


# --- 4. Modelo de Machine Learning (Random Forest Classifier) ---
print("\n--- 4. Modelo de Machine Learning (Random Forest Classifier) ---")

rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

print("Entrenando el modelo Random Forest...")
rf_model.fit(X_train, y_train)
print("Modelo Random Forest entrenado exitosamente.")

y_pred_rf = rf_model.predict(X_test)

print("\nEvaluación del Modelo Random Forest:")
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Precisión (Accuracy): {accuracy_rf:.4f}")

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_rf, target_names=target_names))

conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("\nMatriz de Confusión:")
print(conf_matrix_rf)

plt.figure(figsize=(7, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicción')
plt.ylabel('Etiqueta Verdadera')
plt.title('Matriz de Confusión para Random Forest')
plt.tight_layout()
plt.show()

print("\nImportancia de las Características (Random Forest):")
feature_importances = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)
print(feature_importances)

plt.figure(figsize=(10, len(feature_importances)*0.4 + 2))
sns.barplot(x=feature_importances, y=feature_importances.index, palette='viridis')
plt.title('Importancia de las Características (Random Forest)')
plt.xlabel('Importancia')
plt.ylabel('Característica')
plt.tight_layout()
plt.show()

joblib.dump(rf_model, "modelo_random_forest_salud_mental.pkl")
print("\n Modelo Random Forest guardado como 'modelo_random_forest_salud_mental.pkl'")


# --- 5. Modelo de Deep Learning (Red Neuronal Convolucional 1D - CNN) ---
print("\n--- 5. Modelo de Deep Learning (Red Neuronal Convolucional 1D - CNN) ---")

X_train_cnn = X_train.values.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_cnn = X_test.values.reshape(X_test.shape[0], X_test.shape[1], 1)

print(f"Dimensiones de X_train para CNN: {X_train_cnn.shape}")
print(f"Dimensiones de X_test para CNN: {X_test_cnn.shape}")

print("Construyendo el modelo CNN 1D...")
cnn_model = Sequential([
    Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train_cnn.shape[1], 1)),
    Dropout(0.3),
    Conv1D(filters=128, kernel_size=2, activation='relu'),
    Dropout(0.3),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.4),
    Dense(1, activation='sigmoid') # Capa de salida para CLASIFICACIÓN BINARIA
])

cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # Loss para CLASIFICACIÓN BINARIA
cnn_model.summary()

print("\nEntrenando el modelo CNN 1D...")
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)

history = cnn_model.fit(X_train_cnn, y_train,
                        epochs=200,
                        batch_size=32,
                        validation_split=0.1,
                        callbacks=[early_stopping],
                        verbose=1)

print("Modelo CNN 1D entrenado exitosamente.")

print("\nEvaluación del Modelo CNN 1D:")
loss_cnn, accuracy_cnn = cnn_model.evaluate(X_test_cnn, y_test, verbose=0)
print(f"Pérdida (Loss): {loss_cnn:.4f}")
print(f"Precisión (Accuracy): {accuracy_cnn:.4f}")

y_pred_proba_cnn = cnn_model.predict(X_test_cnn)
y_pred_cnn = (y_pred_proba_cnn > 0.5).astype(int) # Convertir probabilidades a clases binarias (0 o 1)

conf_matrix_cnn = confusion_matrix(y_test, y_pred_cnn)
print("\nMatriz de Confusión (CNN):")
print(conf_matrix_cnn)

print("\nReporte de Clasificación (CNN):")
print(classification_report(y_test, y_pred_cnn, target_names=target_names))

plt.figure(figsize=(7, 6))
sns.heatmap(conf_matrix_cnn, annot=True, fmt='d', cmap='Purples', cbar=False,
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicción')
plt.ylabel('Etiqueta Verdadera')
plt.title('Matriz de Confusión para CNN 1D')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Precisión de Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión de Validación')
plt.title('Precisión del Modelo CNN durante el Entrenamiento')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de Validación')
plt.title('Pérdida del Modelo CNN durante el Entrenamiento')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()


# --- 6. Comparación General de Modelos ---
print("\n--- 6. Comparación General de Modelos ---")
print(f"Precisión (Accuracy) del Random Forest: {accuracy_rf:.4f}")
print(f"Precisión (Accuracy) de la CNN 1D: {accuracy_cnn:.4f}")
print("\nPara una comparación más detallada, revise los Reportes de Clasificación de cada modelo.")
```
